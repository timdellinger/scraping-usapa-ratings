Code that uses BeautifulSoup to open .html files saved from
the USAPA ratings website, and spit out a .csv file for each.
 
 Then read the .csv files back in as a dataframe, delete duplicates,
 and save as a .csv file.  All filenames and folders are hard coded.
  

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python

for i in ['doubles_lvl', 'mixed_lvl', 'singles_lvl']:
    for j in ['2.5','3','3.5','4','4.5','5','5.5','6']:
        print('digesting {} {}'.format(i,j)) 
        
        myfile = r"C:\Tim\pickleball\new ratings system\... pulling all the data Dec 17th\USAPA Pickleball {} {}.html".format(i,j)
        with open(myfile, encoding='utf8') as infile:
            soup = BeautifulSoup(infile, "html.parser")

        my_rows = soup.find("table").find_all('tr')

        data = [[td.findChildren(text=True) for td in tr.findAll("td")] for tr in my_rows]

        headers = [[td.findChildren(text=True) for td in tr.findAll("th")] for tr in my_rows]

        headers_list = headers[0][1:14]

        just_the_headers = [a[0] for a in headers_list]

        df = pd.DataFrame(data)

        df = df.applymap(lambda x: x if not isinstance(x, list) else x[0] if len(x) else '')

        df = df.drop(columns=0)

        df.columns = just_the_headers

        df = df.drop(0)

        df.to_csv(r"C:\Tim\pickleball\new ratings system\... pulling all the data Dec 17th\USAPA Pickleball {} {}.csv".format(i,j))
```

```python
levels = ['2.5','3','3.5','4','4.5','5','5.5','6']
divisions = ['doubles_lvl', 'mixed_lvl', 'singles_lvl']
```

```python
files = ['USAPA Pickleball {} {}.csv'.format(i,j) for i in divisions for j in levels ]
```

```python
files
```

```python

result = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)
```

```python
result = result.drop(result.columns[0],axis=1)
```
