```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python

for i in ['Singles','Doubles','Mixed']:
    for j in ['2.5','3.0','3.5','4.0','4.5','5.0','5.5','6.0']:
        print('hello {} {}'.format(i,j)) 


        soup = BeautifulSoup(open(r"C:\Tim\pickleball\new ratings system\extracting data from 30 sept pull\USAPA Pickleball {} {}.html".format(i,j)), "html.parser")

        my_rows = soup.find("table").find_all('tr')

        data = [[td.findChildren(text=True) for td in tr.findAll("td")] for tr in my_rows]

        headers = [[td.findChildren(text=True) for td in tr.findAll("th")] for tr in my_rows]

        headers_list = headers[0][1:14]

        just_the_headers = [a[0] for a in headers_list]

        df = pd.DataFrame(data)

        df = df.applymap(lambda x: x if not isinstance(x, list) else x[0] if len(x) else '')

        df = df.drop(columns=0)

        df.columns = just_the_headers

        df = df.drop(0)

        df.to_csv(r"C:\Tim\pickleball\new ratings system\extracting data from 30 sept pull\USAPA Pickleball {} {}.csv".format(i,j))
```
